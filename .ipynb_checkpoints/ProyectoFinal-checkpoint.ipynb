#### Librerías a usar

```python
import pandas as pd
import sympy as sp
import math as mt
import numpy as np
import matplotlib.pyplot as plt
```

### Funciones globales

```python
def abrir_csv(entrada):
    """
    Permite abrir un archivo csv y guardarlo en un dataframe (obtejo de Pandas)
    :param
    entrada: Archivo csv que se desee abrir
    :return: Dataframe con los datos del csv. De no encontrarse el archivo o existir otro error, retorna None
    """
    # Intento de lectura del archivo CSV
    try:
        return pd.read_csv(entrada) # Devuelve el dataframe si logró abrirse el csv 
    except FileNotFoundError:  # Archivo no encontrado
        print("Archivo no encontrado:", entrada)  # Mensaje de error
        return None
    except Exception as e:  # Captura de otro tipo de errores
        print("Error al leer el CSV:", e)  # Mensaje de error
        return None
def guardar_csv(df, salida):
    """
    Permite guardar los datos de un dataframe en un archivo csv
    :param
    df: Dataframe que contiene los datos a guardarse en el csv
    salida: csv donde se guardarán los datos
    :return: guarda los datos en el archivo csv, de existir error, no guarda nada y retorna None
    """
    try:
        df.to_csv(salida, index=False)  # Guarda los datos filtrados en un nuevo CSV
    except Exception as e:  # Captura cualquier tipo de error
        print("Error al guardar el archivo.", e)
        return None 
```

### Limpieza del Dataset

#### 1. Eliminación de columnas innecesarias y filtrado de la familia 'BEVERAGES'

```python
datos_entrada = "train.csv"
datos_filtrados = "beverages.csv"

def filtrarDatos(entrada=datos_entrada, salida=datos_filtrados):
    """
    Lee un archivo CSV en busca de las columnas a usar, en caso de encontrarlas y validarlas,
    guarda esas columnas y elimina los registros que no pertenecen a la familia 'BEVERAGES' para
    almacenarlos en un nuevo CSV.
    param:
    entrada: .csv inicial (dataset principal del que se va a filtrar los datos necesarios).
    salida: .csv final (columnas útiles para la ejecución del proyecto).
    :return: Archivo csv con el filtro realizado
    """

    df = abrir_csv(entrada)

    print(f"Registros totales antes de la limpieza: {len(df)}\n")
    
    # Búsqueda y validación de las columnas a usar
    columnasAFiltrar = {"family", "sales", "date"}

    # Valida que las columnas se encuentren en el archivo
    if not columnasAFiltrar.issubset(df.columns):  
        # issubset() se encarga de buscar si cada columna necesaria se encuentra en el archivo
        print("El archivo no contiene las columnas deseadas: ", columnasAFiltrar)  
        return None

    # Busca en la columna "family" todos los registros que tengan el valor "BEVERAGES"
    beverages = df[df["family"] == "BEVERAGES"]  # Filtra únicamente el grupo BEVERAGES
    if "onpromotion" in beverages.columns and "id" in beverages.columns and "store_nbr" in beverages.columns:
        # Busca las columnas onpromotion, id, store_nbr en los datos filtrados beverages
        beverages = beverages.drop(columns = ["onpromotion", "id", "store_nbr"]) # Elimina la columna onpromotion con drop()
    
    # Guardar los datos filtrados en datos_filtrados.csv
    guardar_csv(beverages, salida)

    print(f"Registros restantes después de la limpieza: {len(beverages)}")

    return beverages  # Regresa el dataframe filtrado


# Ejecutar la función
filtrarDatos()
```

#### 2. Construcción de datos semanales

```python
datos_entrada = "beverages.csv"
datos_filtrados = "ventasSemanales.csv"

def agruparVentasSemanales(entrada=datos_entrada, salida=datos_filtrados):
    """
    Lee un CSV con columnas 'date' y 'sales', convierte la fecha a datetime
    y suma las ventas por semana calendario (Lunes a Domingo). Genera un nuevo CSV con las
    ventas semanales y un índice de semana t.

    param:
    entrada: .csv inicial (dataset filtrado previamente con la familia 'BEVERAGES').
    salida: .csv final (datos útiles para la ejecución del proyecto).
    :return: Archivo csv con la agregación realizada
    """
    # Intento de lectura del archivo CSV
    df = abrir_csv(entrada)

    print(f"Registros diarios: {len(df)}\n")

    #Verificar columnas necesarias para la agrupación
    columnasNecesarias = {"date", "sales"}
    if not columnasNecesarias.issubset(df.columns):
        print(f"Error: el archivo no contiene las columnas necesarias {columnasNecesarias} para la agrupación")
        return None

    # Convertir la columna date a datetime
    df["date"] = pd.to_datetime(df["date"], errors="coerce")

    # Asegurar que sales sea una columna numérica
    df["sales"] = pd.to_numeric(df["sales"], errors="coerce")

    # Ordenarmos por fecha
    df = df.sort_values("date")

    # Usamos la fecha como índice para hacer resample por semana
    df = df.set_index("date")

    # Sumamos las ventas por semana 
    ventasSemanales = df["sales"].resample("W-SUN").sum().reset_index()

    #Renombramos la columna date por week_end y sales por sales_week
    ventasSemanales.rename(columns={"date":"week_end","sales":"sales_week"}, inplace = True)

    print(f"Semanas generadas: {len(ventasSemanales)}")

    # Guardar los datos filtrados en ventasSemanales.csv
    guardar_csv(ventasSemanales, salida)

    return ventasSemanales

agruparVentasSemanales()
```

#### 3. Eliminación de posibles valores NaN o registros vacíos

```python
datos_entrada = "ventasSemanales.csv"
datos_filtrados = "datasetFinal.csv"

def eliminacionNan(entrada=datos_entrada, salida=datos_filtrados):
    """
    Lee un archivo CSV, revisa si existen registros con valores NaN, y en caso de encontrarlos,
    elimina esos registros y devuelve un nuevo CSV.

    param:
    entrada: .csv inicial (dataset filtrado previamente con la familia 'BEVERAGES').
    salida: .csv final (datos útiles para la ejecución del proyecto).
    :return: Archivo csv sin registros con valores NaN
    """
    # Intento de lectura del archivo CSV
    df = abrir_csv(entrada)

    print(f"Registros totales antes de la limpieza: {len(df)}\n")

    # Reemplazamos campos vacíos o con espacios por NaN
    dfReemplazado = df.replace('^\s*$', pd.NA, regex=True) # Primer parámetro = qué se va a buscar, en este caso la expresión regular ^\s*$ que significa
    # cero o más espacios en blanco.

    # Eliminar filas con algún NaN
    numRegistrosA = len(dfReemplazado)
    dfLimpio = dfReemplazado.dropna()
    numRegistrosD = len(dfLimpio)

    registrosElim = numRegistrosA - numRegistrosD

    # Mensajes informativos
    if registrosElim == 0:
        print("No se han encontrado registros vacíos o con valores NaN")
        print("El archivo se mantiene igual pero cambia el nombre a 'datasetFinal.csv'")

    print(f"Se eliminaron {registrosElim} registros con valores NaN o vacíos.")
    print(f"Registros restantes después de la limpieza: {numRegistrosD}")

     # Guardar los datos filtrados en eliminadosNan.csv
    guardar_csv(dfLimpio, salida)

    return dfLimpio

eliminacionNan()
```

```python
def graficar_puntos_y_funcion(entrada, f):
    """
    Grafica los puntos de un archivo csv (sin unirlos) y por encima coloca 
    a la función que interpola a dichos datos
    :param
    entrada: Archivo csv que contiene los datos
    f: función que devuelve valores numéricos
    :return: Muestra la gráfica con su función interpolante
    """
    # Cargar datos
    df = abrir_csv(entrada)
    df["week_end"] = pd.to_datetime(df["week_end"])

    # Eje x como índice temporal
    x = np.arange(len(df))
    y = df["sales_week"].values

    # Malla más fina para la función
    x_fino = np.linspace(x.min(), x.max(), 1000)
    y_fino = f(x_fino)

    # Gráfica
    plt.figure()
    plt.plot(df["week_end"], y, '.', label="Datos reales")
    plt.plot(df["week_end"].iloc[x_fino.astype(int)],
             y_fino, label="Modelo")

    plt.title("Ventas vs Semana")
    plt.xlabel("Semana")
    plt.ylabel("Ventas")
    plt.legend()
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
```

#### 2. Estadísticas básicas 

```python
def estadisticas_basicas(entrada, decimales):
    """
    Calcula estadísticas básicas de los datos que se encuentren en el archivo de entrada
    :param
    entrada: archivo csv con los datos de los cuales se calculan las estadísticas
    decimales: número de decimales a mostrar en las estadísticas
    :return: muestra la media, mediana, desviación estándar, mínimo y máximo de los datos
    """
    # Intento de lectura del archivo CSV
    df = abrir_csv(entrada)
    
    # Seleccionar la columna de ventas
    ventas = df["sales_week"]
    
    # Cálculos estadísticos
    media = ventas.mean()
    mediana = ventas.median()
    desviacion_std = ventas.std()
    maximo = ventas.max()
    minimo = ventas.min()
    
    # Mostrar resultados
    print(f"Media: {media:.{decimales}f}")
    print(f"Mediana: {mediana:.{decimales}f}")
    print(f"Desviación estándar: {desviacion_std:.{decimales}f}")
    print(f"Máximo: {maximo:.{decimales}f}")
    print(f"Mínimo: {minimo:.{decimales}f}")
print(f"Estadísticas básicas del dataset final:")
estadisticas_basicas("datasetFinal.csv", 5)
```

#### Separación del dataset

```python
def separar_datos(entrada):
    """
    Separa el archivo con todos los registros agrupados en dos, uno para realizar 
    el ajuste y otro para hacer validaciones con el modelo que se obtenga
    :param 
    entrada: Archivo csv con los datos completos, sin valores NaN y agrupados (datasetFinal.csv)
    :return: Crea dos archivos (datos_ajuste y datos_validacion) csv y muestra cuantos registros contiene cada uno
    """
    # Intento de lectura del archivo CSV
    df = abrir_csv(entrada)
        
    # Número total de registros
    n = len(df)
    
    # Índice de corte (80%)
    corte = int(0.8 * n)
    
    # Separar los conjuntos
    datos_ajuste = df.iloc[:corte]
    datos_validacion = df.iloc[corte:]
    
    # Guardar los nuevos CSV
    guardar_csv(datos_ajuste, "datos_ajuste.csv")
    guardar_csv(datos_validacion, "datos_validacion.csv")
    
    print(f"Total registros: {n}")
    print(f"Registros ajuste (80%): {len(datos_ajuste)}")
    print(f"Registros validación (20%): {len(datos_validacion)}")

separar_datos("datasetFinal.csv")
print("Gráfica de los datos en datos_ajuste.csv")
graficar_puntos("datos_ajuste.csv")
print("Gráfica de los datos en datos_validacion.csv")
graficar_puntos("datos_validacion.csv")
```

### Modelamiento final de la función V(t)

#### Truncamiento de cifras decimales

```python
def truncar(numero, decimales):
    """   
    Permite truncar un número a cierta cantidad de cifras decimales
    :param
     numero: El número a truncar
     decimales: Cuántos decimales mantener
    :return: número truncado en número n de decimales, ejemplo: original= 0.335, 0.33 a 2 decimales truncados
    """
    # Factor multiplicado para mover el punto decimal
    factor = 10 ** decimales
    return mt.trunc(numero * factor) / factor
```

#### Obtener los coeficientes de la función que se obtendrá mediante el método de Mínimos Cuadrados

```python
def coeficientes_minimos_cuadrados_general(func_str, t_lista, y_lista, valores_iniciales):
    """
    Permite encontrar los coeficientes de cualquier modelo que ingrese en el argumento y según los puntos que va a interpolar
    :param 
    func_str: string, función base para el modelado
    t_lista: lista de valores de t
    y_lista: lista de valores de y
    valores_iniciales: lista con valores iniciales, ej: [1, 0.1] (la extensión de la lista debe ser la misma que el número de parámetros)
    :return: diccionario {a: valor_est, b: valor_est, ...}
    """
    # Definicion de la funcion simbolica
    f = sp.sympify(func_str)

    simbolos = list(f.free_symbols) #Obtiene las variables independientes y coeficientes de la función

    # Validacion de la variable independiente
    t = sp.Symbol('t')
    if t not in simbolos:
        raise ValueError("La función debe contener el símbolo 't' como variable independiente.")
    
    parametros = [s for s in simbolos if s != t] #Filtra y guarda únicamente los coeficientes
   
    # Si el numero de valores inicailes no coincide con los parametros da mensaje de error
    if len(parametros) != len(valores_iniciales):
        raise ValueError("El número de valores iniciales debe coincidir con el número de parámetros.")

    S = 0
    #Lógica para las derivadas parciales y resolucion del sistema de ecuaciones
    for ti, yi in zip(t_lista, y_lista):
        S += (yi - f.subs(t, ti))**2

    ecuaciones = [sp.diff(S, p) for p in parametros]

    solucion = sp.nsolve(ecuaciones, parametros, valores_iniciales, tol=1e-8)

    #Crea diccionario con las soluciones
    return {str(parametros[i]): float(solucion[i]) for i in range(len(parametros))} 
```

#### Presentar la función obtenida simbólicamente

```python
def minimos_cuadrados_general_simbolico(func_str, coeficientes, decimales):
    """
    Permite visualizar el modelo obtenido con el valor de sus coeficientes, truncados a cierta cantidad de cifras decimales. 
    El truncamiento es puramente visual, para no agregar error a cálculos posteriores
    :param 
    func_str: string, función original
    coeficientes: diccionario, valor de los coeficientes de la función
    decimales: entero, número de decimales a truncar
    :return: string con la función y sus respectivos coeficientes truncados según decimales
    """
    #Reemplaza los coeficientes de la funcion por sus valores truncados
    for clave, valor in coeficientes.items():
        func_str = func_str.replace(str(clave), str(truncar(valor, decimales)))   
    return func_str
```

#### Obtener la función númerica del modelo

```python
def minimos_cuadrados_general_numerico(func_str, coeficientes):
    """
    Permite crar una función de tipo lambda, para poder darle valores y que devuelva resultados, similar a definir una función matemática en 
    Python usando 'def'
    :param 
    func_str: string, función original
    coeficientes: lista con los coeficientes del modelo
    :return: función lambda, capaz de realizar operaciones del tipo f(arg), donde arg es un valor y f devuelve otro
    """
    # Declclaracion de la variable independiente
    t = sp.Symbol("t")
    # Convercion de la funcion numerica
    func_num = sp.sympify(func_str)
    func_num = func_num.subs(coeficientes)
    return sp.lambdify(t, func_num, "numpy")
```

#### Crear el modelo utilizando `coeficientes_minimos_cuadrados_general()`,`minimos_cuadrados_general_simbolico()` y `minimos_cuadrados_general_numerico()`

```python
def crear_modelo(entrada, modelo_base):
    """
    Permite crear el modelo según los datos de un archivo csv y el modelo base
    :param
    entrada: Archivo csv con los datos para ajustar el modelo
    modelo_base: String con los parámetros y la variable independiente para crear el modelo
    :return: Devuelve tanto el modelo simbólico como numérico
    """
    # Lectura del archivo CSV
    df = abrir_csv(entrada)
    # Bucle for para convertir las columnas a datos numericos
    datos_t = [i + 1 for i in range(len(df))]
    # Lista de valores de y
    datos_y = df['sales_week'].tolist()
    #Obtencion de los coeficientes 
    valores_iniciales = [0] * len([s for s in list(sp.sympify(modelo_base).free_symbols) if s !=sp.Symbol('t')])
    coeficientes = coeficientes_minimos_cuadrados_general(modelo_base, datos_t, datos_y, valores_iniciales)
    
    return minimos_cuadrados_general_numerico(modelo_base, coeficientes), minimos_cuadrados_general_simbolico(modelo_base, coeficientes, 5)
```

```python
v_real, modelo_simbolico = crear_modelo("datos_ajuste.csv", "a*t+b")
print("Modelo de la función V(t) mediante Mínimos Cuadrados")
print(f"{modelo_simbolico}")
```

### Simulación con el modelo

#### Creación de la función

$$f(t) = V_{real}(t) - V_{meta}$$

```python
def f(t, v_meta):
    # Funcion de error entre el modelo y el valor meta
    return v_real(t) - v_meta
```

#### Obtener el intervalo en donde se realizará la predicción

```python
def encontrar_intervalo(v_meta, t_ini=194, t_fin=242, TOL=1e-3):
    """
    Busca intervalo [t, t+1] con cambio de signo para f(t, v_meta),
    pero también detecta si ya se está cerca de la raíz (|f|<TOL).
    :param
    v_meta: meta de ventas que se desea predecir
    t_ini: extremo izquierdo del intervalo donde se busca la raiz, empieza desde 194 que es el dato desde donde empieza datos_validacion.csv
    t_fin: extremo derecho del intervalo donde se busca la raiz, termina en 242 que es el dato donde termina datos_validacion.csv
    TOL: tolerancia
    :return: Si se encuentra, devuelve los valores del intervalo [t, t+1], caso contrario retorna None
    """
    # Evalucacion inicial
    fa = f(t_ini, v_meta)
    # Si el error es menor a la tolerancia en el punto inicial
    if abs(fa) < TOL:
        return (t_ini, t_ini) # Retorna el mismo punto como intervalo
    # Bucle para encontrar el intervalo
    for t in range(t_ini, t_fin):
        fb = f(t + 1, v_meta)
        if abs(fb) < TOL: # Si el error es menor a la tolerancia en el punto final
            return (t + 1, t + 1) # Retorna el mismo punto como intervalo
        if fa * fb < 0: # Cambio de signo
            return (t, t + 1) # Retorna el intervalo encontrado
        fa = fb
    return None
```

#### Método de la Bisección

```python
def biseccion(a, b, v_meta, TOL=1e-3, N_0=100):
    """
    Método de Bisección para hallar t tal que f(t, v_meta)=0
    :param 
    a: punto inicial del intervalo (float/int)
    b: punto final del intervalo (float/int)
    v_meta: meta de ventas (float/int)
    TOL: float, tolerancia
    N_0: int, número máximo de iteraciones
    :return: aproximación p de la raíz, o None si no aplica/fracasa
    """
    # Validacion del intervalo
    if a > b:
        return None
    fa = f(a, v_meta) # Evalua f en a
    fb = f(b, v_meta) # Evalua f en b
    # Si alguno de los extremos es raiz
    if fa == 0:
        return a
    if fb == 0:
        return b
    # Si  no hay cambio de signo no se puede evaluar
    if fa * fb > 0:
        return None # Retorna None
    # Declaracion del contador iterativo
    i = 1
    # Mientras no se alcance el numero maximo de iteraciones
    while i <= N_0:
        # Calculo del punto medio
        p = (a + b) / 2.0
        fp = f(p, v_meta) # Evalua f en el punto medio
        # Si se encuentra la raiz o se cumple la tolerancia
        if fp == 0 or (b - a) / 2.0 < TOL:
            return p # Retorna la aproximacion
        # Si el producto es positivo
        if fa * fp > 0:
            # Actualiza el punto medio como extremo superior
            a = p
            fa = fp
            # Si el producto es negativo
        else:
            # Actualiza el punto medio como extremo inferior
            b = p
            fb = fp
        i += 1
    return None
```

##### Ejemplo de meta de ventas

```python
v_meta = 1362405
intervalo = encontrar_intervalo(v_meta)
if intervalo is None:
    print ("No existe un cambio de signo, no se puede predecir sobre esa meta")
else:
    a, b = intervalo
    t_sol_bi = biseccion(a, b, v_meta, TOL=1e-3, N_0=100)
    print(f"Las semanas en donde se ha encontrado cambio de signo es: [{a},{b}]")
    print(f"Semana estimada (bisección): {t_sol_bi}")
```

### Método de Newton - Raphson

#### Obtener la derivada de la función

```python
def f_derivada(t, v_meta, h=1e-5):
    """
    Aproxima f'(t) con diferencias centrales:
    f'(t) ≈ (f(t+h) - f(t-h)) / (2h)
    :param 
    t: variable independiente
    v_meta: meta de ventas, debe ser la misma que f(t) para que la derivada sea correcta
    h: valor pequeño por el cual se divide, para simular cuando el límite tiende a cero
    :return: 
    """
    # Calculo de la derivada por su definicion
    return (f(t + h, v_meta) - f(t - h, v_meta)) / (2.0 * h)
```

#### Calcular la semana estimada con el método de Newton - Raphson

```python
def newton(p_0, meta, TOL=1e-3, N_0=100):
    """
    Permite encontrar la raiz t tal que f(t, v_meta) = 0
    :param 
    p_0: aproximación inicial
    TOL: float, tolerancia
    N_0: entero, número de iteraciones
    :return: aproximación p de la raíz de la función o None
    """
    # Declaracion del contador iterativo
    i = 1
    i_fallidos = 0 # Contador de fallidos
    # Mientras ni se alcanse el maximo de iteraciones
    while i <= N_0:
        # Aplcacion de la formula 
        p=p_0 - f(p_0, meta)/f_derivada(p_0, meta)
        # Si la derivada es cero no se puede continuar
        if abs(p - p_0)<TOL:
            return p # Retorna la aproximacion
        # Aumenta el contador
        i += 1
        i_fallidos += 1
        p_0 = p # Actualiza el punto inicial
    return None
```

##### Ejemplo de meta de ventas

```python
p_0 = (a + b) / 2
t_sol_ne = newton(p_0, v_meta)
if t_sol_ne is None:
    print ("No se pudo hallar una aproximación para esa meta")
else:
    print(f"Semana estimada (newton): {t_sol_ne}")
```

##### Probar con una lista de metas

```python
def simular_metas(lista_metas, t_ini=194, t_fin=242, tol=1e-3):
    """
    Especie de scrip, permite obtener resultados de varias metas de venta utilizando tanto bisección como Newton-Raphson
    :param
    lista_metas: lista con valores de metas de ventas
    t_ini: extremo izquierdo del intervalo donde se busca la raiz, empieza desde 194 que es el dato desde donde empieza datos_validacion.csv
    t_fin: extremo derecho del intervalo donde se busca la raiz, termina en 242 que es el dato donde termina datos_validacion.csv
    TOL: tolerancia
    :return: Diccionario de diccionarios, cada diccionario interno corresponde a los resultados de una meta de ventas
    """
    # Creacion de diccionario de resultados
    resultados = {}
    # Bucle para cada meta en una lista
    for Vmeta in lista_metas:
        # Busqueda del intervalo
        intervalo = encontrar_intervalo(Vmeta, t_ini=t_ini, t_fin=t_fin)
        if intervalo is None: # Si el intervalo es None
            # Resultados con None
            resultados[Vmeta] = {
                "v_meta": Vmeta,
                "t_bis": None,
                "t_new": None
            }
            continue # Continua con la sigiiente meta
        a, b = intervalo # Retorna el intervalo encontrado
        t_bis = biseccion(a, b, Vmeta, tol) # Aproximacion de la biseccion
        t_new = newton(p_0, Vmeta, tol) # Aproximacion de Newton
        # Guarda los resultados en el diccionario
        resultados[Vmeta] = {
            "v_meta": Vmeta,
            "t_bis": t_bis,
            "t_new": t_new
        }
    return resultados # Retorno del diccionario de resultados
```

```python
meta_venta_agresiva = 1500000
meta_venta_conservadora = 1300000
resultados = simular_metas([meta_venta_agresiva, meta_venta_conservadora])
for keys, values in resultados.items():
    print(values)
```

#### Error de ventas

error_meta = $|v_{real}(round(\hat{t})) - v_{meta}|$

```python
def crear_columna_numeros(df_val, inicio):
    """
    Se crea una columna temporal dentro del dataframe que permite identificar a las columnas de manera secuencial
    :param
    df: Dataframe donde se agregará la columna
    inicio: número desde el que empieza la numeración secuencial
    :return: Dataframe con la columna extra
    """
    # Definicion de la columna week_end como datetime
    df_val["week_end"] = pd.to_datetime(df_val["week_end"], errors="coerce")
    # Definicion del la columna week_num como numeros de la semana
    df_val["week_num"] = np.arange(inicio, inicio + len(df_val))
    return df_val
def error_meta(t_aprox, v_meta):
    """
    Permite calcular el error relativo porcentual comparando la meta de ventas que se alcanza con la raiz de f(t, v_meta) (t_aprox) y el
    valor de la venta real que existe en datos_validacion.csv según t_aprox redondeado
    :param
    t_aprox: Raiz t obtenida de cualquiera de los dos métodos utilizados (bisección, Newton)
    v_meta: meta de ventas con la que se realizará la comparación y cálculo de errores
    :return: error relativo porcentual entre v_real(round(t_aprox)) y v_meta
    """
    # Calculo del error relativo de la meta
    df_val = abrir_csv("datos_validacion.csv") # Lectura del CSV de validacion
    df_val = crear_columna_numeros(df_val, 194)
    t_round = int(round(t_aprox)) # Redondeo de la semana estimada
    venta_real = df_val.loc[df_val["week_num"] == t_round, 'sales_week'].iloc[0] # Valor real de las ventas
    error_meta_abs = abs(venta_real - v_meta) # Calculo del error absoluto
    error_meta_rel = error_meta_abs / abs(v_meta) * 100 # Calculo del error porcentual
    return error_meta_rel
```

#### Calculamos el error relativo porcentual entre las ventas de la semana real y de la semana aproximada

```python
t_bis_agr = resultados[1500000]['t_bis']
print(error_meta(t_bis_agr, 1500000))
t_bis_cons = resultados[1300000]['t_bis']
print(error_meta(t_bis_cons, 1300000))
t_new_agr = resultados[1500000]['t_new']
print(error_meta(t_new_agr, 1500000))
t_new_cons = resultados[1300000]['t_new']
print(error_meta(t_new_cons, 1300000))
```

### MAPE - Mean Absolute Percentage Error

```python
def calcular_mape():
    """
    Permite calcular la métrica MAPE para evaluar al modelo
    :return: valor de MAPE
    """
    # Calculo del MAPE en el conjunto de validacion
    df_val = abrir_csv("datos_validacion.csv") # Lectura del CSV de validacion
    df_val = crear_columna_numeros(df_val, 194)
    errores = [] # Lista para almacenar los errores porcentuales
    # Bucle para calcular el APE de cada punto
    for _, fila in df_val.iterrows():
        t = fila["week_num"] # semana
        y_real = fila["sales_week"] # venta real
        y_hat = v_real(t) # venta estimada por el modelo

        # Evitar división por cero
        if y_real != 0:
            ape = abs(y_real - y_hat) / abs(y_real) * 100
            errores.append(ape)

    # Si no hay errores calculados retorna None
    if len(errores) == 0:
        return None

    return np.mean(errores) # Retorna el MAPE calculado
```

```python
mape = calcular_mape() # Llamada a la funcion de MAPE
print(f"MAPE del modelo en validación: {mape:.4f}%") # Imprime el MAPE redondeado a 4 decimales
```

#### Criterio del 80% de las semanas de validación menores al 20%

```python
def criterio_80_20():
    """
    Permite ver el porcentaje de datos (semanas) dentro de datos_validacion.csv que cumplen con que su APE (Absolute Percentage Error)
    sea menor al 20%. Se considera bueno cuando más del 80% de los datos tiene un APE menor al 20%
    :return: porcentaje de semanas que cumplen con el criterio
    """
    # Lectura del CSV de validacion
    df_val = abrir_csv("datos_validacion.csv") # Lectura del CSV de validacion
    df_val = crear_columna_numeros(df_val, 194)
    cumple = 0 # Contador de puntos que cumplen el criterio
    total = 0 # Contador total de puntos

    # Bucle para calcular el APE de cada punto
    for _, fila in df_val.iterrows():
        t = fila["week_num"]
        y_real = fila["sales_week"]
        y_aprox = v_real(t)
        # Evita la divicion por 0
        if y_real != 0:
            # Calcula el APE y aumenta el contador
            ape = abs(y_real - y_aprox) / abs(y_real) * 100
            total += 1
            # Si el Ape es menor a 20
            if ape < 20:
                cumple += 1 # Actualiza el contador de cumple criterio

    if total == 0: # Si no hay puntos validos retorna None
        return None
    
    porcentaje = cumple / total * 100 # Calculo del porcentaje
    return porcentaje
```

```python
porcentaje_ok = criterio_80_20() # Llamada a la funcion del criterio 80-20
print(f"{porcentaje_ok:.2f}% de las semanas tienen APE < 20%") # Porcentaje redondeado a 2 decimales
```
